```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
consolidado_latino = pd.read_csv('datasets_socioeconomicosF\consolidado_latinoF.csv',sep=',', thousands=',')
canada = pd.read_csv('datasets_socioeconomicosF\canadaF.csv',sep=',', thousands=',')
eeuu = pd.read_csv('datasets_socioeconomicosF\eeuuF.csv',sep=',', thousands=',')
america_latina = pd.read_csv('datasets_socioeconomicosF\\america_latinaF.csv', sep=',', thousands=',')
consolidado_regiones = pd.read_csv("datasets_socioeconomicosF\consolidado_regionesF.csv", sep=',', thousands=',')
```

```python
# Visualizar box plots para identificar outliers
plt.figure(figsize=(10, 6))
sns.boxplot(data=consolidado_regiones["PIB"])
plt.title("Box Plot de PIB")
plt.xticks(rotation=45)
plt.show()
```

```python
columns_to_plot = consolidado_regiones.columns[2:]
# Ploteo de histogramas para cada columna
for column in columns_to_plot:
    plt.figure(figsize=(8, 6))
    plt.hist(consolidado_regiones[column], bins=50, edgecolor='black')
    plt.title(f"Histograma de {column}")
    plt.xlabel(column)
    plt.ylabel("Frecuencia")
    plt.show()
```

```python
columns_to_plot = consolidado_regiones.columns[2:]
for column in columns_to_plot:
    plt.figure()
    
    if column in eeuu.columns:
        plt.hist(eeuu[column], bins=20, edgecolor='black')
        plt.title(f"Histograma de {column} - EEUU")
        plt.xlabel(column)
        plt.ylabel("Frecuencia")
        plt.show()
    
    if column in canada.columns:
        plt.hist(canada[column], bins=20, edgecolor='black')
        plt.title(f"Histograma de {column} - Canad")
        plt.xlabel(column)
        plt.ylabel("Frecuencia")
        plt.show()
    
    if column in america_latina.columns:
        plt.hist(america_latina[column], bins=20, edgecolor='black')
        plt.title(f"Histograma de {column} - Amrica Latina")
        plt.xlabel(column)
        plt.ylabel("Frecuencia")
        plt.show()
```

```python
# Ploteo de diagramas de dispersin con colores por categora de regin
import seaborn as sns
import matplotlib.pyplot as plt

sns.pairplot(consolidado_regiones, hue='Pais', diag_kind='kde')
plt.suptitle("Diagramas de Dispersin de Indicadores Socioeconmicos por Regin", y=1.02)
plt.show()
```

```python
# Aos seleccionados
years_selected = [1990, 1995, 2000, 2005, 2010, 2015, 2020]

# Filtrar el DataFrame para incluir solo los aos seleccionados
df_years = consolidado_regiones[consolidado_regiones['Ao'].isin(years_selected)]
```

```python
# Lista de nombres de indicadores en tu DataFrame consolidado_regiones
indicadores = ["PIB", "Pobreza", "Tasa de empleo", "Mortalidad"]

# Ciclo for para generar grficos para cada indicador
for indicador in indicadores:
    # Agrupar por ao y pas y obtener la suma del indicador
    grouped_data = df_years.groupby(["Ao", "Pais"])[indicador].sum().unstack()

    # Crear grfico de barras apiladas
    ax = grouped_data.plot(kind="bar", stacked=True)

    # Agregar ttulo y etiquetas
    plt.title(f"{indicador} por regin a lo largo de los aos")
    plt.xlabel("Ao")
    plt.ylabel(indicador)

    plt.show()
```

```python
# Eliminar espacios en blanco de la columna 'Pais'
consolidado_latino['Pais'] = consolidado_latino['Pais'].str.strip()
```

```python
# Aos seleccionados
years_selected = [1990, 1995, 2000, 2005, 2010, 2015, 2020]

# Filtrar el DataFrame para incluir solo los aos seleccionados
df_years2 = consolidado_latino[consolidado_latino['Ao'].isin(years_selected)]
```

```python
# Lista de pases de Amrica del Sur
paises_suramerica = ["Argentina", "Bolivia (Estado Plurinacional de)", "Brasil", "Chile", "Colombia", "Ecuador", "Guyana", "Paraguay", "Per", "Suriname", "Uruguay", "Venezuela (Repblica Bolivariana de)"]

# Crear un DataFrame filtrando los pases de Amrica del Sur
df_suramerica = df_years2[df_years2['Pais'].isin(paises_suramerica)]
```

```python
# Lista de nombres de indicadores en tu DataFrame df_suramerica
indicadores = ["PIB", "Pobreza", "Tasa de empleo", "Mortalidad", "Alfabetizacion", "Aos de estudio"]

# Ciclo for para generar grficos para cada indicador solo para Amrica del Sur
for indicador in indicadores:
    # Agrupar por ao y pas y obtener la suma del indicador
    grouped_data = df_suramerica.groupby(["Ao", "Pais"])[indicador].sum().unstack()

    # Crear grfico de barras apiladas
    ax = grouped_data.plot(kind="bar", stacked=True,legend = False)

    # Agregar ttulo y etiquetas
    plt.title(f"{indicador} en Amrica del Sur a lo largo de los aos")
    plt.xlabel("Ao")
    plt.ylabel(indicador)

    # Crear una leyenda desplegable
    ax.legend(labels=grouped_data.columns, title="Pases", bbox_to_anchor=(1.05, 1), loc='upper left')
    

    plt.show()
```

```python
# Ciclo for para generar grficos de lneas para cada indicador solo para Amrica del Sur
for indicador in indicadores:
    # Agrupar por ao y pas y obtener la suma del indicador
    grouped_data = df_suramerica.groupby(["Ao", "Pais"])[indicador].sum().unstack()

    # Crear grfico de lneas
    ax = grouped_data.plot(kind="line", legend= False)

    # Agregar ttulo y etiquetas
    plt.title(f"{indicador} en Amrica del Sur a lo largo de los aos")
    plt.xlabel("Ao")
    plt.ylabel(indicador)
    # Crear una leyenda desplegable
    ax.legend(labels=grouped_data.columns, title="Pases", bbox_to_anchor=(1.05, 1), loc='upper left')
    
    plt.show()
```
